```python
class Gaussain_T():
    def __init__(self,mu,std):
        self.mu = mu
        self.std = std
        self.num_samples = 100
        
    def get_samples(self):
        samples = np.random.normal(self.mu, self.std,self.num_samples)
        return samples
```
