# Coin flip simulator and distributions

**Coin flip simulator**

```python
import numpy as np
import random
import scipy.stats as stat 
import matplotlib.pyplot as plt
```

```python

class Coin_simulator:
    
    def __init__(self,no_of_experiments):
        self.experiments = no_of_experiments
        
    def coin_flip(self):
        head_count = 0
        tail_count = 0
        # all possible faces
        faces = ['T', 'H'] 
        for i in range(0,self.experiments):
            # randomly choose a face
            top_face = random.choice(faces) 
            
            #checking if we got head face
            if top_face == 'H':
                head_count +=1 
            else:
                tail_count +=1

        return [head_count,tail_count]
    
    def plot(self):
        count = coin_flip()
        heads = count[0]
        tails = count[1]
        print("frquency of heads :",heads)
        print("frquency of tails :",tails)
        bar=plt.bar(["Heads"," Tails"],[heads,tails] )
        plt.show()

Coin_simulator(100).plot()
```

# Discrete Distributions

**Binomial Distribution**

```python
binomial=stat.binom(n=10,p=0.4)
t=np.arange(0,11)
binomial_sample=binomial.rvs(size=1000,random_state=143)
plt.hist(binomial_sample,bins=11,range=(0,11),density=True,align='left',rwidth=0.75,label="Sample",alpha=0.75)
plt.title("Binomial Distribution")
plt.legend()
plt.xticks(t)
plt.show()
```

**Possion Distribution**

```python
poisson = stat.poisson(mu=2)
poisson_sample = poisson.rvs(size=1000,random_state=143)
end = np.max(poisson_sample)+1
t=np.arange(0,end)
plt.hist(poisson_sample, bins=end,range=(0,end), density=True,align='left', rwidth=0.75,label="Sample", alpha=0.75)
plt.title("Poisson Distribution")
plt.legend()
plt.show()
```

# **Continuous Distribution**

**Normal Distributions**

```python
normal=stat.norm(loc=100, scale=15)
normal_sample = normal.rvs(size=1000,random_state=123)
plt.hist(normal_sample,density=True,label="Sample")
plt.title("Normal Distribution")
plt.legend()
plt.show()
```

**Beta Distribution**

```python
beta=stat.beta(a=2,b=5)
beta_sample=beta.rvs(size=1000,random_state=133)
t=np.linspace(0,1,100)
plt.hist(beta_sample,density=True,label="Sample")
plt.legend()
plt.title("Beta Distribution")
plt.show()
```

# **Multivariant Distributions**

**Multivariant Normal Distribution**

```python
x = np.linspace(0, 5, 10, endpoint=False)
y = stat.multivariate_normal.pdf(x, mean=2.5, cov=0.5); y
plt.plot(x, y)
```

**Multivariant Nominal Distribution**

```python
n = 5
p=[10]
random_var = stat.multinomial(n, p)

pmfs = list()
xs = list()
for x in range(n-4, n+5, 1):
    pmfs.append(random_var.pmf([x]))
    xs.append(x)

plt.plot(xs, pmfs)
```
