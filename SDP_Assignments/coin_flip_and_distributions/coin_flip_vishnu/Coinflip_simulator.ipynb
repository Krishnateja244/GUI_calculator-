```python
import numpy as np
import random
import scipy.stats as stat 
import matplotlib.pyplot as plt
```

```python
number_of_flips = 100

def flip_coin(number_of_flips):
    '''
    Function to flip a coin n times.
    
    Parameters: 
    ----------------
    number_of_flips - [int]
    
    Returns:
    --------------
    coin_flip_results - [list]
    '''
    coin_flip_results=[]
    for i in range(0,number_of_flips):
        coin_flip_results.append(random.randint(0,1))
    return coin_flip_results

coin_flip_list = flip_coin(number_of_flips)
heads = coin_flip_list.count(1)
tails = coin_flip_list.count(0)
print("Number of heads: ", heads)
print("Number of tails: ", tails)
bar=plt.bar(["Heads"," Tails"],[heads,tails] )
bar[0].set_hatch('/')
bar[1].set_hatch('o')
plt.ylabel("Count")
plt.show()
```

# Discrete probability distributions

In probability, a discrete distribution has either a finite or a countably infinite number of possible values.

Poisson distribution:
---------------------

A Poisson distribution is the probability distribution that results from a Poisson experiment.  A Poisson experiment is a statistical experiment that classifies the experiment into two categories, such as success or failure.  A Poisson random variable “x” defines the number of successes in the experiment. This distribution occurs when there are events that do not occur as the outcomes of a definite number of outcomes

$p(x) = \frac{e^{-\lambda} * \lambda^{x}}{x !}$

x is a Poisson random variable.

λ is an average rate of value.


```python
lam = 25
x_vals = np.linspace(0,100,100)
poisson_ditribution = stat.poisson.pmf(x_vals,mu=lam)
plt.plot(poisson_ditribution)
plt.xlabel("Random variable")
plt.ylabel("pmf")
plt.show()
```

Bionomial distribution:
------------------------
The binomial distribution formula helps to check the probability of getting “x” successes in “n” independent trials of a binomial experiment. The probability distribution becomes a binomial probability distribution when it meets the following requirements:

Each trail can have only two outcomes or the outcomes that can be reduced to two outcomes

These outcomes can be either a success or a failure.

The trails must be a fixed number.

The outcome of each trial must be independent of each other.

And the success of probability must remain the same for each trail.

$p(x) = n_{C_{r}} . p^{r} . (1-p)^{(n-r)}$


```python
prob = 0.5

distibution = []
for i in range(0,len(x_vals)):
    k = stat.binom.pmf(i,number_of_flips,prob)
    distibution.append(k)

plt.bar(x_vals,distibution)
plt.xlabel("Random variable")
plt.ylabel("pmf")
plt.show()
```

# Continuos probability distributions

If the variable associated with the distribution is continuous, then such a distribution is said to be continuous.

Normal Distribution:
--------------------

Normal distribution is also known as Gaussiann distribution. The Normal Distribution is defined by the probability density function for a continuous random variable in a system. The probability density function of normal or gaussian distribution is given by:

$P(x) = \frac{1}{\sigma \sqrt{2\pi}}e^{\frac {-(x- \mu)^{2}}{2 \sigma^{2}}}$

x = random variable

$\mu$ = mean

$\sigma$ = standard deviation

```python
mean = 55
std = 5
var = std**2
samples =50000 

def gauss_dist(mean,var,x):
    gauss = (1/(var*np.sqrt(2*np.pi)))*np.exp(-((x-mean)**2)/(2*(var**2)))
    return gauss

random_varible_x = np.linspace(0,100,100)
gauss_dist_val=[]
for i in random_varible_x:
    x=i
    y=gauss_dist(mean,var,x)
    gauss_dist_val.append(y)

sample_distribution = np.random.normal(mean,var,samples)
plt.plot(random_varible_x,gauss_dist_val,linewidth=2, color='Green')
plt.hist(sample_distribution, 200, density=True)
plt.legend(["Normal distribution","samples"])
plt.xlabel("Random variable")
plt.ylabel("pdf")
plt.show()
```

Exponential Distribution:
-------------------------

The exponential distribution is a probability distribution which represents the time between events in a Poisson process. It is also called negative exponential distribution. It is a continuous probability distribution used to represent the time we need to wait before a given event happens. 

The probability density function (pdf) of an exponential distribution is given by:

$p(x) = \lambda.e^{-\lambda.x}$

```python
exponential_distribution = stat.expon.pdf(x_vals)
plt.plot(exponential_distribution)
plt.xlabel("Random variable")
plt.ylabel("pdf")
plt.show()
```
