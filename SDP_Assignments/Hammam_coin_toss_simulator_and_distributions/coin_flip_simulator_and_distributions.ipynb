```python
import random    
import matplotlib.pyplot as plt 
from scipy import stats as st
import numpy as np 
%matplotlib inline 
```

```python
class Coin_Flip:
    
    def __init__(self,num_of_experiments):
        self.num_of_experiments = num_of_experiments 

                
    def results(self):        
        count_heads = count_tails = 0
        for event in range(self.num_of_experiments):         
            sample = random.randint(0,1)
            if sample == 1:
                count_tails += 1
            else: 
                count_heads += 1 
        return [count_heads,count_tails]
    
    
    def plot_dist(self):
        dist = self.results()
        plt.bar(["Heads","Tails"],[dist[0],dist[1]])       
        plt.ylabel("Count")
        plt.title("Coint flip distribution for {} flips".format(self.num_of_experiments))
        plt.grid()
        plt.show()
                
        
```

```python
Coin_Flip(10).plot_dist()
```

```python
Coin_Flip(100).plot_dist()
```

```python
Coin_Flip(1000).plot_dist()
```

```python
Coin_Flip(10000).plot_dist()
```

# Continious Distributions 

## Rayleigh Distribution 

```python
x = np.linspace(0,10) 
y = st.rayleigh.pdf(x)
plt.plot(x,y)
plt.xlabel("Random Variable")
plt.ylabel("PDF")
plt.title("Rayleigh Distribution")
plt.grid()
plt.show()
```

# Uniform Distribution

```python
x = np.linspace(-3,3) 
y = st.uniform.pdf(x)
plt.plot(x,y)
plt.xlabel("Random Variable")
plt.ylabel("PDF")
plt.title("Uniform Distribution")
plt.grid()
plt.show()
```

# Discrete Distributions

## Bernoulli Distribution 

```python
prob = 0.4
x = np.linspace(-1,2,4) 
y = st.bernoulli.pmf(x,prob)
plt.bar(x,y)
plt.xlabel("Random Variable")
plt.ylabel("PMF")
plt.title("Bernoulli Distribution")
plt.grid()
plt.show()
```

```python
prob = 0.4
n = 20
x = np.linspace(-1,2,4) 
y = st.binom.pmf(x,n,prob)
plt.bar(x,y)
plt.xlabel("Random Variable")
plt.ylabel("PMF")
plt.title("Bernoulli Distribution")
plt.grid()
plt.show()
```

## Multivariate Normal Distribution 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
#Parameters to set
mu_x = 0
variance_x = 1

mu_y = 0
variance_y = 3

#Create grid and multivariate normal
x = np.linspace(-5,5,500)
y = np.linspace(-5,5,500)
X, Y = np.meshgrid(x,y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y
rv = multivariate_normal([mu_x, mu_y], [[variance_x, 0], [0, variance_y]])
Z = rv.pdf(pos)
#Make a 3D plot
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X, Y, Z,cmap='viridis',linewidth=0)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
plt.show()
```
